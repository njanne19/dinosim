```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
beamwidth_3dB = 65
gain_3dB = 3
K = beamwidth_3dB * np.sqrt(gain_3dB)
```

```python
def gaussian_radiation_pattern(gain, K): 
    beamwidth = K / np.sqrt(gain) 
    sigma = np.radians(beamwidth) / 2
    theta = np.linspace(-np.pi, np.pi, 360) 
    pattern = gain * np.exp(-0.5 * (theta / sigma) ** 2)
    pattern_dB = 10 * np.log10(pattern)
    pattern_normalized_dB = pattern_dB - max(pattern_dB)

    return theta, pattern_normalized_dB 

plt.figure()
for i, gain in enumerate([3, 5, 8, 12], start=1): 
    theta, pattern = gaussian_radiation_pattern(gain, K) 
    ax = plt.subplot(2, 2, i, polar=True) 
    ax.plot(theta, pattern) 
    ax.set_title(f"Gain: {gain}dB") 
    # Removed setting ylim automatically adjust for Gaussian's max value 
    ax.set_ylim(-40, 0)
    ax.grid(True) 

plt.tight_layout()
plt.show() 
```
